```python
Q1. Explain the difference between simple linear regression and multiple linear regression. Provide an example of each.
Simple Linear Regression:
In simple linear regression, there is one independent variable predicting a dependent variable.
Example: Predicting a student's final exam score (dependent variable) based on the number of hours they studied (independent variable).
Multiple Linear Regression:
In multiple linear regression, there are two or more independent variables predicting a dependent variable.
Example: Predicting a house's price (dependent variable) based on square footage, number of bedrooms, and location (independent variables).


Q2. Discuss the assumptions of linear regression. How can you check whether these assumptions hold in a given dataset?
Assumptions:
Linearity: The relationship between variables is linear.
Independence: Residuals are independent.
Homoscedasticity: Residuals have constant variance.
Normality: Residuals are normally distributed.
Checking Assumptions:
Residual plots: Check for patterns in residual plots to assess linearity and homoscedasticity.
Durbin-Watson test: Assess independence of residuals.
Normal probability plot: Check for normality in residuals.


Q3. How do you interpret the slope and intercept in a linear regression model? Provide an example using a real-world scenario.
Slope (Coefficient): Represents the change in the dependent variable for a one-unit change in the independent variable.
Intercept: Represents the predicted value of the dependent variable when all independent variables are zero.
Example:
In a simple linear regression predicting salary based on years of experience:
Slope: A slope of $5,000 means that, on average, each additional year of experience is associated with a $5,000 increase in salary.
Intercept: An intercept of $40,000 means that a person with zero years of experience is estimated to have a starting salary of $40,000.


Q4. Explain the concept of gradient descent. How is it used in machine learning?
Gradient Descent:
Gradient descent is an optimization algorithm used to minimize the cost function in machine learning models.
It iteratively adjusts model parameters (weights) in the direction of the steepest decrease in the cost function.
Process:
Initialize weights randomly.
Calculate the gradient (partial derivatives) of the cost function.
Update weights by moving in the opposite direction of the gradient.
Repeat until convergence.


Q5. Describe the multiple linear regression model. How does it differ from simple linear regression?
Multiple Linear Regression:
In multiple linear regression, there are two or more independent variables predicting a dependent variable.
Difference from Simple Linear Regression:
Simple linear regression has only one independent variable, while multiple linear regression has multiple.


Q6. Explain the concept of multicollinearity in multiple linear regression. How can you detect and address this issue?
Multicollinearity:
Multicollinearity occurs when independent variables in a multiple linear regression model are highly correlated.
It can lead to unstable coefficient estimates and challenges in interpreting the individual contribution of variables.
Detection and Addressing:
Detection: Calculate variance inflation factor (VIF). High VIF values indicate multicollinearity.
Addressing: Remove or combine highly correlated variables, or use regularization techniques like ridge regression.


Q7. Describe the polynomial regression model. How is it different from linear regression?
Polynomial Regression:
Polynomial regression is an extension of linear regression where the relationship between variables is modeled as an nth-degree polynomial.
Difference from Linear Regression:
Linear regression models linear relationships, while polynomial regression can capture non-linear relationships by introducing polynomial terms.


Q8. What are the advantages and disadvantages of polynomial regression compared to linear regression? In what situations would you prefer to use polynomial regression?
Advantages:
Flexibility: Can model non-linear relationships.
Accuracy: Provides a better fit for complex relationships.
Disadvantages:
Overfitting: Higher-degree polynomials may overfit the data.
Interpretability: Coefficients are harder to interpret.
Use Cases:
Situation: When the relationship between variables is clearly non-linear.
Example: Predicting stock prices where the influence of factors may not be linear.
Situation: When a higher degree of accuracy is required, and overfitting can be controlled.
Example: Predicting sales based on marketing spend, considering non-linear effects.
```
